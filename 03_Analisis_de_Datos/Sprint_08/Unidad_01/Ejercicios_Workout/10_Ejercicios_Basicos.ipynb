### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
# Crear los datos
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Crear la figura y los subplots
plt.figure(figsize=(10, 5))

# Primer subplot
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primer subplot
plt.plot(x, y_sin)
plt.title('Seno')

# Segundo subplot
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segundo subplot
plt.plot(x, y_cos)
plt.title('Coseno')

# Mostrar la figura
plt.tight_layout()
plt.show()
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
# Crear los datos
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

# Crear la figura y los subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 10))

# Primer subplot: Seno
ax1.plot(x, y_sin, color='red')
ax1.set_title('Seno')

# Segundo subplot: Coseno
ax2.plot(x, y_cos, color='green')
ax2.set_title('Coseno')

# Tercer subplot: Tangente
ax3.plot(x, y_tan, color='blue')
ax3.set_title('Tangente')
ax3.set_ylim(-40, 40)  # Limitar el eje y para evitar valores extremos

# Ajustar el diseño
plt.tight_layout()
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Línea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
# Crear los datos
x = np.linspace(0, 10, 100)
y_sin = np.sin(2 * x)
y_cos = np.cos(2 * x)

# Crear la figura y los subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Primer subplot: Seno con estilo magenta discontinua
ax1.plot(x, y_sin, 'm--')
ax1.set_title('Seno')

# Segundo subplot: Coseno con estilo cian guiones y puntos
ax2.plot(x, y_cos, 'c-.')
ax2.set_title('Coseno')

# Ajustar el diseño
plt.tight_layout()
plt.show()
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperbólico y coseno hiperbólico se obtienen así:
* np.sinh, seno hiperbólico;
* np.cosh, coseno hiperbólico;


```python
# Data for plotting
x = np.linspace(0, 10, 100)
sinh_x = np.sinh(x)
cosh_x = np.cosh(x)

# Create the plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Seno Hiperbólico
ax1.plot(x, sinh_x, 'r')
ax1.set_title('Seno Hiperbólico')
ax1.set_ylabel('sinh(x)')
ax1.set_xlabel('x')

# Coseno Hiperbólico
ax2.plot(x, cosh_x, 'b')
ax2.set_title('Coseno Hiperbólico')
ax2.set_ylabel('cosh(x)')
ax2.set_xlabel('x')

# Show plot
plt.tight_layout()
plt.show()

```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python
# Data for plotting
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log1p(x)  # log1p(x) is ln(1 + x)

# Create the plots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Plot each function in a subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_xlim(0, 10)
axs[0, 0].set_ylim(-1, 1)

axs[0, 1].plot(x, y2)
axs[0, 1].set_xlim(0, 10)
axs[0, 1].set_ylim(-1, 1)

axs[1, 0].plot(x, y3)
axs[1, 0].set_xlim(0, 10)
axs[1, 0].set_ylim(-40, 20)

axs[1, 1].plot(x, y4)
axs[1, 1].set_xlim(0, 10)
axs[1, 1].set_ylim(0, 1.5)

# Adjust spacing to prevent overlap
plt.subplots_adjust(hspace=0.5, wspace=0.5)

# Show plot
plt.show()

```


### Ejercicio 6: 

Reproduce la figura siguiente empleando sólo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros gráficos (fila 1). Luego usa el estilo con métodos (objetos `axes`) para añadir el tercer gráfico.

```python
# Data for plotting
x = np.linspace(0, 10, 100)
cos_x = np.cos(x)
sqrt_x = np.sqrt(x)
sin_x = np.sin(x)

# Create the figure and subplots using MATLAB style
plt.figure(figsize=(8, 8))

# First subplot (top)
plt.subplot(2, 1, 1)
plt.plot(x, cos_x, label='coseno')
plt.plot(x, sqrt_x, label='raiz cuadrada')
plt.title('Raíz Cuadrada y coseno')
plt.legend()

# Second subplot (bottom)
plt.subplot(2, 1, 2)
plt.plot(x, sin_x)
plt.title('Seno')

# Show plot
plt.tight_layout()
plt.show()

# Create the third subplot using the object-oriented style
fig, ax = plt.subplots(1, 1, figsize=(8, 4))

ax.plot(x, sin_x)
ax.set_title('Seno')

plt.show()
```
